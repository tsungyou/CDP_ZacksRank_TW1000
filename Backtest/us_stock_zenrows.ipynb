```python
import requests

def get_zacks_data(symbol):
    url = f"https://quote-feed.zacks.com/index?t={symbol}"
    response = requests.get(url)
    
    if response.status_code == 200:
        return response.json()  # Assuming the response is in JSON format
    else:
        print(f"Failed to fetch data for {symbol}. Status code: {response.status_code}")
        return None

# Example usage
symbol = 'TSLA'
data = get_zacks_data(symbol)
print(data)
```

```python
from selenium import webdriver
import pandas as pd
import time 
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import datetime
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def next_weekday(time_var):
    weekday = time_var.weekday()
    if weekday in [6, 0, 1, 2, 3]:
        escape = 1
    elif weekday == 4:
        escape = 3
    elif weekday == 5:
        escape = 2
    time_var = time_var + datetime.timedelta(days=escape)
    return time_var

driver = webdriver.Firefox()
driver.get("https://www.nasdaq.com/market-activity/earnings")


# codes = driver.find_elements(By.TAG_NAME, "a")
# print(codes)
time_var = datetime.datetime.now()

df = {}
for a in range(4):
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    date_button = soup.find_all("button", {"class":"time-belt__item"})
    page_count = soup.find_all("button", {"class":"pagination__page"})
    for b in range(1, len(page_count)):
            
        s = soup.find_all("td", {"class":"market-calendar-table__cell market-calendar-table__cell--cell"})

        

        for c in range(0, len(s)-1, 8):
            # print(i)
            k = s[c].text.strip()
            df[k] = time_var.strftime("%Y-%m-%d")
        if b != len(page_count)-1:
            next_page_button_xpath = f'//button[@data-page="{b}"]'
            
            # Wait for the element to be clickable
            next_page_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, next_page_button_xpath))
            )
            
            # Click the element using JavaScript to avoid visibility issues
            driver.execute_script("arguments[0].click();", next_page_button)

    if a != 3:
        for d in date_button:
            print(d.text)
        
        time_var = next_weekday(time_var)
        year = time_var.year
        month = "0" + str(time_var.month)
        date = time_var.day
        wait = WebDriverWait(driver, 10)
        element = wait.until(EC.presence_of_element_located((By.XPATH, f'button.time-belt__item[data-year="{year}"][data-month="{month}"][data-day="{date}"]'), timeout=20))
        element.click()



# dfs = pd.DataFrame([df]).T.reset_index()
# dfs.columns = ['Date', 'Value']
# dfs
# dfs = pd.DataFrame(df.keys(), index=df.values())
# dfs.columns=["code"]
# dfs
```
